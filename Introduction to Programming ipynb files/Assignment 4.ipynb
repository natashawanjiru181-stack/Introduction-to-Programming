```python
#Name: Natasha Gicheha
# 1. Common Friends Finder
#Two friends mention the people they know
# Friend 1 
Susan = {'Ceaser', 'Faith', 'Sydney', 'Joy', 'Wanjiru', 'Sam'} #Friend one writes people she knows
print(Susan)
```

```python
#Friend 2
#Friends she knows
Angela = {'Faith', 'Mark', 'Joy', 'Miriam', 'Njoroge'} #Friend 2 mentions people she knows
print(Angela)
```

```python
#People that both friends know
# 1a. Use & operator
both = Susan & Angela #Both friends of Susan and Angela
print(both)
```

```python
# Name: Natasha Gicheha
# 1b. Names that are unique to friend 1
#Use - operator
unique = Susan - Angela #Friends of Susan only
print(unique)
```

```python
#Set of all unique friends known by either friend
# 1c. Use | operator
all_unique = Susan | Angela #all unique friends
print(all_unique)
```

```python
# 2. Two siblings have to do chores
#Each writes their set of chores
Max = {'taking out the garbage', 'cooking',  'lawn mowing', 'repairing the broken water pipes'} #set of 1st sibling's chores
print(Max)
```

```python
# Name: Natasha Gicheha
# 2nd sibling chores
Chloe = {'cooking', 'cleaning the house', 'laundry', 'taking out the garbage'} #set of 2nd sibling's chores
print(Chloe)
```

```python
 #Chores common in both siblings
# 2a. Use & operator
common = Max & Chloe #Both Max and Chloe do these chores
print(common)
```

```python
# 2b. Chores that only one sibling wrote down
#Symmetric difference
#Use ^ operator
only_one = Max ^ Chloe #Chores that only Max does
print(only_one)
```

```python
## Name: Natasha Gicheha
# 2c. Different chores Max and Chloe were doing
diff1 = Max - Chloe #Chores that Max was doing but Chloe wasn't
print(diff1)

diff2 = Chloe - Max #Chores that Chloe was doing but Max wasn't
print(diff2)
```

```python
# 2d. Combine the two differences
nobody = diff1 | diff2 #Chores that nobody shares
print(nobody)
```

```python
# Name: Natasha Gicheha
# 3. Unique words counter
mantra = 'Chocolate is heavenly Chocolate is heavenly why not' #Create a sentence to be used
print(mantra)
```

```python
# 3a. Split the sentence into words
split = mantra.split() #Sentence is split
print(split)
```

```python
# 3b. Store unique words in a set
set = set(split) #Set of unique words
print(set)
```

```python
#Name: Natasha Gicheha 
# 3c. Total unique words
total = len(set) #How many unique words are found in the sentence
print(total)
```

```python
# 3d. Show set of unique words
print(set) #Output is unordered
```

```python
# 4. Real world mini problem
#Problems affecting the youth today
#Data from 1st student
Zoe={'Unemployment', 'Drug abuse', 'Corruption', 'Depression', 'Cancer', 'HIV/AIDS'} #Problems affecting 1st student
print(Zoe)
```

```python
#Name: Natasha Gicheha
# Problems affecting 2nd student
#Data from 2nd student
Millie={'Unemployment', 'Depression', 'Truancy', 'Laziness', 'Corruption'} #Problems affecting 2nd student
print(Millie)
```

```python
# 4a. Find unique problems of 1st student
#Use - operation
Zoe_unique = Zoe - Millie #Problems only Zoe is facing
print(Zoe_unique)
```

```python
# 4b. Find common problems in the students
#Use & operation
common = Zoe & Millie #Problems in both students
print(common)
```

```python
#Name: Natasha Gicheha 
# 4c. Get unique problems
#Use ^ operation
unique_prob = Zoe ^ Millie #Unique problems of both students
print(unique_prob)
```

```python
#Difference between Millie and Zoe
Millie_unique = Millie - Zoe #Unique problems of Millie only
print(Millie_unique)
```

```python
# 4d. Unique problems of both students
set_union = Zoe_unique | Millie_unique #All unique problems
print(set_union)
```
