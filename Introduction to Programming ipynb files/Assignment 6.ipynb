```python
#Assignment 6
#if, elif and else statements

#Task 1: Even or odd
test=int(input("Is number even or odd?"))
print(test)#The number is 98

#If number is even
if test % 2 ==0:
    print("Number is even")
    
#If number is false
else:
    print("Number is odd")
```

```python
#Task 2: Age Group
age=int(input("What is your age?"))
print(age) #Age is 98
```

```python
#Print whether the age mentioned above falls in the following categories
#Below 13 is a child
if age<13:
    print("Child")
    
#Between 13 and 19 years old
elif 13>= age >=19:
    print("Teenager")
    
#If between 20 and 64 years old
elif 20>= age >=64:
    print("Adult")
    
#If this person is 65 years old or older
if age>=65:
    print("Senior")
```

```python
#Task 3: Maximum of three numbers
#User enters 3 numbers which fall under x, y, g variables
x=input("First number?:")# is 78
print(x)

y=input("Second number?:")# is 48
print(y)

g=input("Third number?:")# is 65
print(g)
```

```python
#Comparing these numbers using nested ifs
if x>=y and x>=g: #Goal is to identify which number is the largest
    print("The largest number is 78")
else:
    print("The largest number is 48")
```

```python
#Different method of printing the largest number
print("The largest number is",max(x,y,g))
```

```python
#Task 4: Grading System
#Prompt a random student's score and print it
student=int(input("Student's score?:"))
print(f"{student}%")
```

```python
# Task 4
#Print whether the student's grade is in these boundaries
#Grade A is from 80 to 100
if 80<=student<=100:
    print("A")
    
#Grade B is from 60 to 79
if 60<=student<=79:
    print("B")
    
#Grade C range is 50 to 59
elif 50<=student<=59:
    print("C")
    
#Grade D is from 40 to 49 marks
elif 40<=student<=49:
    print("D")
    
#Any mark below 40 is an F grade
elif student<40:
     print("F")
     
#If student's grade is not within the above categories, else statement applies
else:
    print("Invalid")
```

```python
#Task 5: Temperature Check
temp=float(input("Give me any temperature ")) #Mention a temperature value
print(f"{temp} \u00B0C")#\u00B0C to show temperature units, degree Celsius
```

```python
#Print messages based on temperature value provided above 
#Temperature below 15 degrees Celsius only
if temp<15:
    print("Cold")
   
#Temperature between 15 degrees Celsius and 30 degrees Celsius 
elif 15<=temp<=30:
    print("Warm")
 
#Temperature above 30 degrees Celsius only   
if temp>30:
    print("Hot")
    
#If temperature value does not fall in the above categories, else statement applies
else:
    print("Invalid")
```
