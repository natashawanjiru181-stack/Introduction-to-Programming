```python
#Line plot
import matplotlib.pyplot as plt
x=[5,6,7,8,9] #x axis
y=[10,2,3,4,5] #y axis
plt.plot(x,y)
plt.show()
```

```python
#Bar plot
import matplotlib.pyplot as plt
x=[5,6,7,8,9] #x axis
y=[10,2,3,4,5] #y axis
plt.bar(x,y)
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt
y=[2,3,4,5] #y axis
plt.hist(y)
```

```python
#Scatter plot
import matplotlib.pyplot as plt
x=[5,6,7,8,9] #x axis
y=[3,2,3,4,5] #y axis
plt.scatter(x,y)
plt.show()
```

```python
#Visualizing without using x and y axis
import numpy as np
plt.title('Line graph using functional approach')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
x=np.linspace(0,10,20)#20 datapoints between 0 and 10
y=x**2 #array y from square of x
plt.scatter(x,y)
plt.show()
```

```python
#Create 2 plots on the same canvas
plt.subplot(1,2,1)
plt.plot(x,y, 'pink')#Only one color
plt.title('Pink Line Graph')

plt.subplot(1,2,2)
plt.plot(y,x, 'orange');
plt.title('Orange Line Graph')
```

```python
#Object Oriented Interface
#Blank figure
fig=plt.figure()
a=fig.add_axes([0.1, 0.2, 0.3, 0.7])
a.plot(x,x**2, 'green', label='X Plot')
a.plot(x,x**2, 'purple', label='Y Plot')
a.legend()
```

```python
#Plot inside a plot
fig=plt.figure()
a=fig.add_axes([0.1, 0.1, 0.7, 0.7])
ab=fig.add_axes([0.3, 0.2, 0.5, 0.4])

a.plot(x,y)
ab.plot(y,x)
```

```python
#Matrix subplot(3*3matrix)
fig, axes=plt.subplots(nrows=3, ncols=3)
axes[0,1].plot(x,y)
axes[1,2].plot(y,x)
plt.tight_layout()
```
