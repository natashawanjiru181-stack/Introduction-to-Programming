```python
# Loops
# Types of loops: while loop, for loop
#  Exercises: Level 1
### 1) Iterate **0 to 10** using a `for` loop and a `while` loop

# Using for loop
for i in range(11):
    print(i)

# Using while loop
i = 0
while i <= 10:
    print(i)
    i += 1
```

```python
# 2) Iterate **10 to 0** using a `for` loop and a `while` loop
# Using for loop
for i in range(10, -1, -1):
    print(i)

# Using while loop
i = 10
while i >= 0:
    print(i)
    i -= 1
```

```python
# 3) Print a # triangle (7 rows)
for i in range(1, 8):
    print('#' * i)
```

```python
#4) Use nested loops to print an 8Ã—8 block of # characters

for _ in range(8):           # 8 rows
    row = []
    for _ in range(8):       # 8 columns
        row.append('#')
    print(' '.join(row))
```

```python
#5) Print: i x i = i*i for i from 0 to 10
for i in range(11):
    print(f"{i} x {i} = {i*i}")
```

```python
#6) Iterate a list using a for loop

tech_list = ['Python', 'Numpy','Pandas','Django', 'Flask']
for item in tech_list:
    print(item)
```

```python
# Exercises: Level 2
#9) Sum of all numbers from 0 to 100

total = 0
for i in range(101):
    total += i
print(f"The sum of all numbers is {total}.")
```

```python
# 10) Sum of all evens and odds from 0 to 100

sum_evens = 0
sum_odds = 0
for i in range(101):
    if i % 2 == 0:
        sum_evens += i
    else:
        sum_odds += i
print(f"The sum of all evens is {sum_evens}. And the sum of all odds is {sum_odds}.")
```

```python

import ast

path = r"C:\Users\T2Gic\Downloads\countries_data.py"  # keep .py if you want
with open(path, "r", encoding="utf-8") as f:
    text = f.read().strip()

countries_data = ast.literal_eval(text)  # safely parse Python literals into objects
print(type(countries_data), len(countries_data))
```

```python
#C) Work with countries_data.py
from collections import Counter

#i) total number of languages in the data
all_langs = []
for c in countries_data:
    all_langs.extend(c.get('languages', []))

unique_langs = set(all_langs)
print("Total number of languages:", len(unique_langs))

```

```python
#ii) Find the ten most spoken languages from the data
lang_counts = Counter(all_langs)
top_10_langs = lang_counts.most_common(10)
print("Ten most spoken languages (language, count):")
for lang, count in top_10_langs:
    print(f"{lang}: {count}")
```

```python
#iii) Find the 10 most populated countries in the world
sorted_by_pop = sorted(
    countries_data,
    key=lambda c: c.get('population', 0),
    reverse=True
)
top_10_pop = sorted_by_pop[:10]
print("Top 10 most populated countries (name, population)")
for c in top_10_pop:
    print(f"{c['name']}: {c['population']}")
```
