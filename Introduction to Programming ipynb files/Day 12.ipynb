```python
#Exercises: Level 1
import random
import string

# 1) Generate a six-character random user ID
def random_user_id():
    # Use letters + digits, pick 6 random characters
    chars = string.ascii_letters + string.digits
    return ''.join(random.choice(chars) for _ in range(6))

print(random_user_id())  # Example: '1ee33d'
```

```python
# 2) Generate multiple user IDs based on user input
def user_id_gen_by_user():
    # Ask user for number of characters and number of IDs
    chars_per_id = int(input("Enter number of characters per ID: "))
    num_ids = int(input("Enter number of IDs to generate: "))
    chars = string.ascii_letters + string.digits
    ids = []
    for _ in range(num_ids):
        ids.append(''.join(random.choice(chars) for _ in range(chars_per_id)))
    return '\n'.join(ids)

# Example usage:
# print(user_id_gen_by_user())  # Input: 5 and 5
```

```python
# 3) Generate a random RGB color
def rgb_color_gen():
    # Random integers between 0 and 255 for R, G, B
    r, g, b = (random.randint(0, 255) for _ in range(3))
    return f"rgb({r},{g},{b})"

print(rgb_color_gen())  # Example: rgb(125,244,255)
```

```python
#Exercises: Level 2

# 4) Generate a list of random HEX colors
def list_of_hexa_colors(n):
    # Each color is '#' followed by 6 hex digits
    colors = []
    for _ in range(n):
        color = '#' + ''.join(random.choice('0123456789abcdef') for _ in range(6))
        colors.append(color)
    return colors

print(list_of_hexa_colors(3))  # Example: ['#a3e12f','#03ed55','#eb3d2b']
```

```python
# 5) Generate a list of random RGB colors
def list_of_rgb_colors(n):
    return [rgb_color_gen() for _ in range(n)]

print(list_of_rgb_colors(3))  # Example: ['rgb(5,55,175)','rgb(50,105,100)','rgb(15,26,80)']
```

```python
# 6) Generate colors based on type ('hexa' or 'rgb')
def generate_colors(color_type, count):
    color_type = color_type.lower()
    if color_type == 'hexa':
        return list_of_hexa_colors(count)
    elif color_type == 'rgb':
        return list_of_rgb_colors(count)
    else:
        return "Invalid color type"

print(generate_colors('hexa', 3))
print(generate_colors('rgb', 3))
```

```python
#Exercises: Level 3

# 7) Shuffle a list
def shuffle_list(lst):
    shuffled = lst[:]
    random.shuffle(shuffled)
    return shuffled

print(shuffle_list([1, 2, 3, 4, 5]))  # Example: [3, 1, 5, 2, 4]
```
