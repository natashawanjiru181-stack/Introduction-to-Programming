```python
# 2. Create a tuple 
# containing names of your sisters and your brothers
sisters=('Farida', 'Fiona')
brothers=('Ryan', 'Max')

print(f'Sisters={sisters}')
print(f'Brothers={brothers}')
```

```python
# 3. Join brothers and sisters tuples and assign it to siblings
# i)Convert the tuples into lists for it to be modified
a=list(sisters)
b=list(brothers)

print(a)
print(b)
```

```python
# ii) Join brothers and sisters tuples and assign it to siblings list
siblings=a+b
print(f'Siblings={siblings}')
```

```python
# iii) Convert the siblings list to a tuple
print(tuple(siblings))
```

```python
# 4. How many siblings do you have?
# Use len function
a=len(siblings)
print(f'{a} siblings')
```

```python
# 5. Modify the siblings tuple
# i) Convert the siblings tuple to a list 
a=list(siblings)
print(a)
```

```python
# ii) add the name of your father and mother 
# using append function
a.append('James')
a.append('Caroline')
print(a)
```

```python
# iii) and assign it to family_members
family_members=a
print(family_members)
```

```python
# Level 2
# 1. Unpack siblings and parents from family_members
*siblings, parent1, parent2= family_members
print('Parents:', f"{parent1} and {parent2}")
print('Siblings:', ','.join(siblings[:-1]) + " and " + siblings[-1]) 
```

```python
# 2. Create fruits, vegetables and animal products tuples. 
fruits=('Mangoes', 'Watermelons', 'Oranges', 'Passion')
vegetables=('Kales', 'Cabbages', 'Onions', 'Tomatoes')
animal_product=('Hides', 'Milk', 'Beef', 'Pork')

print(fruits)
print(vegetables)
print(animal_product)
```

```python
# i) Join the three tuples and assign it to a variable called food_stuff_tp.
food_stuff_tp= fruits + vegetables + animal_product
print(food_stuff_tp)
```

```python
# 3. Change the tuple to a food_stuff_lt list
food_stuff_lt = list(food_stuff_tp)
print(food_stuff_lt)
```

```python
# 4. Slice out the middle item
# i) Find the middle of the list
a=len(food_stuff_lt)//2
print(a)# middle of the list is found at index 6
```

```python
# ii) Slice out the middle item
a=food_stuff_lt[6]
print(a)
```

```python
# 5. Slice out the first three items
a=food_stuff_lt[:3]
print(a)
```

```python
# i) Slice out the last three items
a=food_stuff_lt[-3:]
print(a)
```

```python
# 6. Delete food_stuff_tp Tuples
del food_stuff_tp #success

#Test if it worked
print(food_stuff_tp)
```
