```python
#Personal string variable
HOT_STUFF='Your beauty cannot be described'
print(HOT_STUFF)
```

```python
#Number of characters in string variable
len(HOT_STUFF)
```

```python
#Slicing in Python
#1st 5 characters
#Use curved brackets
ANGEL=HOT_STUFF[:5]
print(ANGEL)
```

```python
#last 5 characters
MINE=HOT_STUFF[5:]
print(MINE)
```

```python
#Get characters from index 2 - index 7
BOMB_SHELL=HOT_STUFF[2:7]
print(BOMB_SHELL)
```

```python
#Reverse the whole string
CUTEY=HOT_STUFF[::-1]
print(CUTEY)
```
