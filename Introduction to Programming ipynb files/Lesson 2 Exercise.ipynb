```python
#Area of a circle
#math.pi=pie

import math

#radius=30m
area_of_circle=math.pi * 30**2
print(f'{area_of_circle:.2f}m\u00B2')

#Type of variable
type(area_of_circle)
```

```python
#Circumference of a circle
circum_of_circle=math.pi * (30*2)

#Round off the answer to 2 decimal places
print(f'{circum_of_circle:.2f}m')

#Type of variable used
type(circum_of_circle)
```
