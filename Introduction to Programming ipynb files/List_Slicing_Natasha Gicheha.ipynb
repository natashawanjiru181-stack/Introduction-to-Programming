```python
#Task 1
#Creating a list
#Should contain a mix of numbers and strings
shopping=['cookies',20, 'cream',40, 'milk',70, 'jam',98, 'sugar',76, 'honey',25, 'syrup',87]
print(shopping)
```

```python
#Number of elements in list
#Should be 12 elements minimum
len(shopping)
```

```python
#Task 2
#Basic slicing, my shopping list
#print 1st 5 elements only/slice last elements
violet=shopping[:5]
print(violet)
```

```python
#Slicing
#Last 5 elements
purple=shopping[-5:]
print(purple)
```

```python
#Slicing
#Elements from index 3 - index 8
red=shopping[3:8]
print(red)
```

```python
#Task 3
#Step Slicing, printing every 2nd element
#Starting from index 0
green=shopping[::2]
print(green)
```

```python
#Slicing
#Every 3rd element, starting from index 2
orange=shopping[2::]
print(orange)
```

```python
#Reversing shopping list using slicing
yellow=shopping[::-1]
print(yellow)
```

```python
#Task 4
#Negative indexing with slicing
#Last 7 elements using negative indices
pink=(shopping[-7:])
print(pink)
```

```python
#Every 2nd element starting at the end
black=(shopping[::-2])
print(black)
```

```python
#Slicing
#Exclude;1st element and last 2 elements
white=(shopping[1:-2])
print(white)
```

```python
#Task 5
#Practice challenge
#Slicing with Step
grey=(shopping[0:13:3])
print(grey)
```
