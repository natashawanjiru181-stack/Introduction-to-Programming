```python
#Using a 'for' loop in lists
#To collect each student's score
scores=[]#main list, 1st create it while empty
for i in range(1,15): #Loops from one to 15
    score=int(input('Score for each student?:'))#Enter an individual's score
    scores.append(score)#enables the list to change as a value is being added
print(f'{scores}%')
```

```python
#Using a for loop in strings
mantra=input('What do you feel like telling yourself right now?:')
vowels='aeiouAEIOU'#mention all the vowels in uppercase and lowercase letters
count=0

for char in mantra:
    if char in vowels:
        count += 1
print(f'Total vowels: {count}')
```

```python
#Using a while loop
#Validating each score is between 0 and 100
while score < 0 or score > 100:#if statement was true it would never stop looping
    print('Really low score')
else:
    print('Everyone scored highly')
```

```python
#Print the highest score
print(f'The Highest Score is {max(scores)}%')
```

```python
#Average score
av=sum(scores)/len(scores)
print(f'The Average Score is {av:.2f}%')
```

```python
#List of scores above 75%
above_75=[score for score in scores if score > 75]
print(above_75)# 8 students have above 75%
```

```python
#Create a list of numbers
num=[30, 76, 30, 76, 4, 56, 12, 12, 9, 8, 9, 8, 5, 64]
print(num)
```
