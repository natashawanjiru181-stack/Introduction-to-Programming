```python
#Read into csv
#Two methods
# 1. Using DictReader
import csv
with open('CreateCSVfile.csv', mode='r')as csvfile:
    m=list(csv.DictReader(csvfile))
    print(m)
```

```python
#Read into csv
# 2. Using reader
import csv
with open('CreateCSVfile.csv', mode='r')as csvfile:
    m=list(csv.reader(csvfile))
    print(m)#Output is divided into 2, categories and values- for 1st row
```

```python
#Write into csv using Distwriter
import csv
with open('CreateCSVfile2.csv',mode='w',newline='')as csvfile:
    fieldnames=['name', 'age', 'major']
    writer=csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({'name': 'Gichy Ansh', 'age': 56, 'major': 'IST'})
    writer.writerow({'name': 'Dean Hurt', 'age': 34, 'major': 'IST'})
    
```

```python
#Write into csv using Distwriter
#Creating a new csv file
import csv
with open('CHI Quiz.csv',mode='w',newline='')as csvfile:
    fieldnames=['Country(ENG)', 'Country(CHI)', 'Celebrity']
    writer=csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    
    #Number of rows and content in each 
    #Each value if non numerical must be in quotes
    #6 rows, 3 key-value pairs(Dictionaries)
    writer.writerow({'Country(ENG)': 'United States', 'Country(CHI)': 'Meiguo', 'Celebrity': 'Kim Kardashian'})
    writer.writerow({'Country(ENG)': 'Argentina', 'Country(CHI)': 'Agenting', 'Celebrity': 'Kylian Mbappe\''})
    writer.writerow({'Country(ENG)': 'Portugal', 'Country(CHI)': 'Putaoya', 'Celebrity': 'Cristiano Ronaldo'})
    writer.writerow({'Country(ENG)': 'Canada', 'Country(CHI)': 'Jianada', 'Celebrity': 'Justin Bieber'})
    writer.writerow({'Country(ENG)': 'United States', 'Country(CHI)': 'Meiguo', 'Celebrity': 'Barack Obama'})
    writer.writerow({'Country(ENG)': 'Kenya', 'Country(CHI)': 'Kenniya', 'Celebrity': 'Rachel Ruto'})
```
