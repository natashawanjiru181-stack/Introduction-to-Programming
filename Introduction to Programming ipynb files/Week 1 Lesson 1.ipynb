```python
#Slicing strings
#Slicing always ignores the first index mentioned
#Slicing is done in square brackets
b='Hello, World!'
s=b[2:5] #ignored character in index 2, counted from index 3 to index 5
print(s)
```

```python
#Slicing strings 
#To get the first 5 characters
#Method A do it in a variable
b='Hello, World!'
c=b[0:5] 
print(c)

#Method B do it in print
b='Hello, World!'
print(b[0:5]) #expected output Hello
```

```python
#Method C
#Get the first 5 characters
d='OrangeBigger'
e=d[:6]# ignore index 6
print(e)
```

```python
#Last 5 characters - success
a='When in need, reach out'

#Step 1 get total length
print(len(a))

#Step 2- total length 23 - 5 =18
print(a[18:])# index 18 is ignored, output has 5 characters including spacing
```

```python
#Get charaters from index 2 - index 7
a='When in need, reach out'
b=a[1:7] # index 1 is ignored starts counting from index 2
print(b)# output has 6 characters
```

```python
#Reverse string using slicing
a='My, World'
b=a[::-1]
print(b)# whole string is reversed
```

```python
#Replacing letters in a string
# using replace function
a = 'Wonderful, World'
b = a.replace('W', 'M') #replace letter W with M
print(b)
```
