```python
# Check for full row duplicates
a=df.duplicated().sum()
print(f'Full row duplicates={a}')
```

```python
# Check for full column duplicates
# 1. Define the function
def get_duplicate_columns(df):
    duplicates = []
    for i in range(df.shape[1]):
        col1 = df.iloc[:, i]
        for j in range(i + 1, df.shape[1]):
            col2 = df.iloc[:, j]
            if col1.equals(col2):
                duplicates.append((df.columns[i], df.columns[j]))
    return duplicates

#Check full column duplicates
a=get_duplicate_columns(df)
print(f'Full column duplicates={a}')#Empty/no full column duplicates
```
