```python
#Create a list
list=[1,2,3,4,5,6]
print(list)
```

```python
#10 was put in the second index
#10 replaced 3 in the list
#Start counting indexes from zero
list[2]=10;
print(list)
```

```python
#Adding multiple element
list[1:4]=[34,67]
print(list)
```

```python
#Using append method
list.append(45)
print(list)
```

```python
#Insert method
#Specify where you want it
#Replaced index 2
list.insert(2,'awesome')
print(list)
```

```python
#Removing items from a list
#Use remove method
list.remove('awesome')
print(list)
```

```python
#Pop method removes the specific index
#Value in index 3 was removed
list.pop(3)
print(list)
```

```python
#Using del to remove items
#If you don't speify del function will delete everything
del list[0]
print(list)
```

```python
#Using tuples
#Create tuple
#Order of items in tuple are unchangeable 
# Allow duplicate items
tuple=('apple', 'orange', 'mango',)
print(tuple)
```

```python
#Length of my tuple
len(tuple)
```

```python
#Access tuple items
print(tuple[2]) #Picked value in second index
#Remember start counting indexes from 0
```

```python
#Negative indexing
#Starts counting from the end
```

```python
#Range of indexes
tuple=('apple', 'orange', 'mango', 'melon', 'kiwi', 'lemon')
print(tuple[2:5])
```

```python
#Remove an item
#Showing elements from the start but doesnt include the index specified
print(tuple[:3])
```
