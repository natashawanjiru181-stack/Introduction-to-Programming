```python
#Dictionaries are ordered
#Stores values in key value pairs
#Uses curly brackets
# 1. Create a dictionary
#Key is brand, value is Mercedez
dict={'brand':'Mercedez', 'model':'Discovery', 'year':'1453'}
print(dict)
```

```python
#Print a value from a certain key from the dictionary
print(dict['brand'])
```

```python
#Dictionaries are changeable
#Can't have 2 items with the same key
#Duplicate values overwrite existing values
# 3. Test theory
dict={'brand':'Mercedez', 'model':'Discovery', 'year':'1453', 'year':'3091'}#ist year was ignored
print(dict)
```

```python
#Length of a dictionary
len(dict) #three key value pairs
```

```python
#Items in dictionary can be of any data type
dict={'brand':'Ford', 'electric':'false', 'year':'1967', 'color':['pink', 'orange', 'green']}
E=dict['color'] #print the values of color key
print(E)
```

```python
#Get keys
#keys()method returns a list of all the keys in the dictionary
dict={'brand':'Ford', 'electric':'false', 'year':'1967', 'color':['pink', 'orange', 'green']}
x=dict.keys()
print(x) #before the change
```

```python
#After the change
dict['age'] = 'new'# adding a new key and value pair
print(x)
```

```python
#Get values
#Shows all the values in the dict only
dict={'brand':'Ford', 'electric':'false', 'year':'1967', 'color':['pink', 'orange', 'green']}

x=dict.values()
print(x)#before
```

```python
#Change one of the values
dict['brand']='Isuzu'
print(x)
```

```python
#Update dictionary
#Updates the dict with items from another dict or key value pairs
dict={'brand':'Ford', 'electric':'false', 'year':'1967', 'color':['pink', 'orange', 'green']}

#Use update function
dict.update({'electric':'true'})
print(dict)
```

```python
#Copy a dictionary
#Use copy() method
dict={'brand':'Ford', 'electric':'false', 'year':'1967', 'color':['pink', 'orange', 'green']}
newdict=dict.copy()
print(dict)
```
