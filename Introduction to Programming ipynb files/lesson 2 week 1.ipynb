```python
#Slicing strings
b='Hello, World!'
s=b[2:5]#Starts counting from index 2 to index 5
print(s)
```

```python
#Slicing strings to get 
# the first 5 characters
yooh='''
Spicy
Burger
'''
daa=yooh[:5]
print(daa)#4 characters are visible, 5th character is a new line
```

```python
#Slicing strings 
#To get the first 5 characters
d='OrangesareBigger'
e=d[:5]
print(d)#total of 16 characters
len(d)
```

```python
#Last 5 characters
a='When in need,reach out'#The string can have spaces, it doesn't affect the answer
kay=(a[5:])#Starts counting from index 5 all the way to the last index
print(kay)
```

```python
#Testing slicing of a string
r90='Watermelon Sugar High'

#Slice from index 6 towards the end
pro=r90[6:]
print(pro)
```

```python
#Get charaters from index 2 - index 7
a='When in need, reach out'
b=a[2:7]
print(b)#starts at index 2 all the way to index 7
```
