```python
#Removing items from a tuple
# 1st create the tuple
tuple=('apple', 'orange', 'pineapple')
print(tuple)
```

```python
#2nd change the tuple to a list
change=list(tuple)
print(change)
```

```python
#Remove an item from the list
change.remove('orange')
```

```python
#Python sets
#create a set
#uses curly brackets
python_set={'apple', 'banana', 'mango'}
print(python_set)
```

```python
#test if 
#sets allow duplicates
python_set={'apple', 'banana', 'mango', 'apple'} #duplicate was ignored
print(python_set)
```

```python
#set items-data types
#string
set = {'mango', 'passion', 'apple'}
#integers
set1 = {1,2,3,4,5}
#boolean
set2 = {True, False, True}
```

```python
#adding items in a set
#cannot change items in a set but you can add
#values in a set are unordered
set.add('pear')
print(set)
```

```python
#add items from another set
set2.update(set1)
print(set2)
```

```python
#removing items from a set
#using remove() method
set1.remove(5)
print(set1)
```

```python
#removing items from a set
#discard()method
#if item to remove isn't there discard() won't function
set1.discard(2)
print(set1)
```

```python
#pop method
#removes items randomly
#don't specify
set2.pop()
print(set2) #removed false element
```

```python
#using clear method
set2.clear() #this method clears the set
print(set2)
```
