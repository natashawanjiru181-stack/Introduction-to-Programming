```python
#Python while loops
i=1
while i<6:
      print(i)
      i+=1#this is the most important part
```

```python
#Break statement
#stops the loop even if while condition is true
i=1
while i<6:#i=3
      print(i)
      if i==3:
          break
      i+=1
```

```python
#Continue statement
#Stops the current iteration and continues with the next
i=0
while i<6:
    i+=1 #Add one to i
    if i==3:
        continue
    print(i)
```

```python
#Else statement
#Runs the code when condition is not true
i=1
while i<6:
      print(i)
      i+=1
else:
    print('i is no longer less than 6')
      
```

```python
#Linear search algorithm
list=[4,3,5,6,7]
s=5 #value i'm searching for. 8 is not in the list
i=0 #start index for the loop
while i<len(list):
    if list[i]==s:
        print('found',list[i])# 5 was found in the list
        break
    i+=1
else:
    print('no match')
```

```python
#Python for loops
#used in lists, strings, tuples, dictionary, set
fruits=['apple', 'banana', 'orange']
for x in fruits:#x takes each value in the list one by one, vertically
    print(x)
#if i used print instead of for, values would appear horizontally
```

```python
#Loop through letters
for x in 'orange':#letters are printed vertically
    print(x)
```

```python
#Break statement
#to stop the loop before it has looped through all items
fruits=['apple', 'banana', 'orange']
for x in fruits:#x takes each value in the list one by one, vertically
    print(x)
    if x=='banana':
     break# break statement appears after print
```

```python
# break statement appears before print
fruits=['apple', 'banana', 'orange']
for x in fruits:#x takes each value in the list one by one, vertically
    if x=='banana':
        break
    print(x)
    
```
